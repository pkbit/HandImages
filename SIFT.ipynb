```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance
```

```python
img = cv2.imread('/Users/satrajitmaitra/Downloads/Hands/Hand_0011742.jpg')
img2 = cv2.imread('/Users/satrajitmaitra/Downloads/Hands/Hand_0011744.jpg')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
```

```python
orb = cv2.ORB_create()
```

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp, des = sift.detectAndCompute(gray,None)
kp2, des2 = sift.detectAndCompute(gray2,None)
```

```python
matches = []
for vec in des:
    min_dist = distance.euclidean(vec, des2[0])
    for i in range(1, len(des2)):
        dist = distance.euclidean(vec, des2[i])
#         print(dist)
        if min_dist > dist:
            min_dist = dist
    matches.append(min_dist)
sum(matches)
```

```python
bf = cv2.BFMatcher(cv2.NORM_L1,crossCheck=False)
```

```python
matches = bf.match(des,des2)
matches = sorted(matches, key = lambda x:x.distance)
len(matches)
```

```python
sum([match.distance for match in matches])
```

```python
kp = sift.detect(gray,None)

img=cv2.drawKeypoints(gray,kp,gray)
# cv2.imshow('ORB keypoints', img)
plt.imshow(img,cmap='gray')
plt.show()
```
